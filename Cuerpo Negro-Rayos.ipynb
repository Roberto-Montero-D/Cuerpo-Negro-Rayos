```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math

def circle():
    r = random.uniform(0.05,0.3)
    x,y = random.uniform(r,1.0-r),random.uniform(r,1.0-r)
    
    return x,y,r

def ray():
    x_r = np.arange(0, 1, 0.001)
    b = random.uniform(-1,2)
    m = random.uniform(-1.0,1.0)
    y_r = m*x_r + b
    return x_r, y_r

count = 0        
x,y,r = circle()
figure, axes = plt.subplots()
color = (random.random(), random.random(), random.random())
circle = plt.Circle((x, y), r, color=color)
axes.add_artist(circle)
axes.set_xlim([0, 1])
axes.set_ylim([0, 1])
axes.set_aspect(1) 
plt.show()

print('x =',x)
print('y =',y)
print('r =',r)

```

```python
figure, axes = plt.subplots()
axes.set_xlim([0, 1])
axes.set_ylim([0, 1])
axes.set_aspect(1)
pos_x = []
pos_y = []
for i in range(1000):
    x_r , y_r = ray()
    d_list=[]
    for i in range(len(x_r)):
        d_list.append(math.sqrt((x_r[i]-x)**2+(y_r[i]-y)**2))
    hit = 0
    
    for i in range(len(d_list)):
        
        if d_list[i] < r:
            hit +=1
            pos_x.append(x_r[i])
            pos_y.append(y_r[i])
            
    if hit !=0:
        count+=1
    plt.plot(x_r, y_r)
plt.show()




```

```python
figure, axes = plt.subplots()
axes.set_xlim([0, 1])
axes.set_ylim([0, 1])
axes.set_aspect(1)
plt.scatter(pos_x, pos_y, s=1)
plt.show()
```

```python
print('Propiedades objeto:')
print('x =',x)
print('y =',y)
print('r =', r)
print('')


min_x = min(pos_x)
max_x = max(pos_x)
min_y = min(pos_y)
max_y = max(pos_y)
r_aprox_x = (max_x-min_x)/2
r_aprox_y = (max_y-min_y)/2
x_aprox = (min_x+max_x)/2
y_aprox = (min_y+max_y)/2



print('Aproximaciones:')
print('r_aprox_x =',r_aprox_x)
print('r_aprox_y =',r_aprox_y)
print('x_aprox =',x_aprox)
print('y_aprox =',y_aprox)

print('')
print('Optimizacion y errores:')
diff_x = abs(r - r_aprox_x)
diff_y = abs(r - r_aprox_y)
if min(diff_x,diff_y) == diff_x:
    mejor = r_aprox_x
    error_r = diff_x / r
else:
    mejor = r_aprox_y
    error_r = diff_y / r
    
error_x = abs(x-x_aprox)/x
error_y = abs(y-y_aprox)/y

print('La mejor aproximacion de r es', mejor)
print('El porcentaje de error para el radio es %.2f' %(error_r*100),'%')

print('El porcentaje de error para la posicion x es %.2f' %(error_x*100),'%')
print('El porcentaje de error para la posicion y es %.2f' %(error_y*100),'%')


```
